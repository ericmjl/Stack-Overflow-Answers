```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
data
```

```python
grouped = data.groupby(['Buyer', 'Seller']).sum().reset_index().set_index('Buyer')
grouped
```

```python
grouped['Total'] = data.groupby('Buyer').sum()['Amount']
grouped
```

```python
grouped['Share'] = grouped['Amount'] / grouped['Total']
grouped
```
